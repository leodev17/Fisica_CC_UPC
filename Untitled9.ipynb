```python
import matplotlib.pyplot as plt
from numpy import sqrt
import numpy as np
#Refraccion de la luz
#Vamos a considerar dos medios distintos separados cuya interfase es el eje x.
# Luego, tomaremos el punto P1(x1,y1) en el medio 1 y sabemos que un haz de luz que partió desde P1 pasó también
# por el punto P2(x2,y2). Según el modelo teórico, las trayectorias que puede tomar el haz de luz consisten
# en un segmento de recta de P1 a la interfase y luego otro segmento de recta desde ese punto final hasta P2.
# La trayectoria tomada dependerá de en qué punto de la interfase pasa el haz de luz de tal forma que esta sea
# la trayectoria que tome el menor tiempo teniendo en cuenta que la luz tiene distintas velocidades en ambos medios.
# Defincion de parametros:
x1=2
y1=6
x2=6
y2=-7
n1=1.33 #El medio superior sera el agua
n2=2.42 #El medio inferior sera vidrio Pyrex
c=300000000 #Velocidad de la luz en m/s
v1=c/n1
v2=c/n2
def tiempo_trayectoria(x):
    return sqrt((x1-x)**2+y1**2)/v1 + sqrt((x2-x)**2+y2**2)/v2
f_vect=np.vectorize(tiempo_trayectoria)
arr_x=np.arange(0,10,0.001)
arr_t=f_vect(arr_x)
indice_minimo=np.argmin(arr_t)
x_interf=arr_x[indice_minimo]
sin2=(x2-x_interf)/sqrt((x2-x_interf)**2+y2**2)
sin1=(x_interf-x1)/sqrt((x1-x_interf)**2+y1**2)
ang_2=np.degrees(np.arcsin(sin2))
ang_1=np.degrees(np.arcsin(sin1))
plt.title("Gráfica Tiempo vs x")
plt.xlabel("Distancia x en la que la luz intersecta a la interfase (m)")
plt.ylabel("Tiempo (s)")
plt.plot(arr_x, arr_t)
plt.show()
print("sen(a1)/sen(a2) = " + str(sin1/sin2))
print("n2/n1 = " + str(n2/n1))
print("a1 = " + str(ang_1))
print("a2 = " + str(ang_2))
```
